연속형 변수인 기온, 풍속, 습도, 강수량, 전력사용량에 대해 boxplot을 그려본다.



<기온>

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots()

기온 = train['기온(°C)']

ax.boxplot([기온])
ax.set_ylim(0.0, 40.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```

<풍속>

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots()
풍속 = train['풍속(m/s)']

ax.boxplot([풍속])
ax.set_ylim(-10.0, 25.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

plt.show()
```
